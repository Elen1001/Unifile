# 4. Տվյալների մաքրում

```python
import numpy as np

# Թվային սյուներ
num_cols = df.select_dtypes(include=np.number).columns

# Լրացնում ենք բաց թողնված թվային արժեքները միջինով
for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mean(), inplace=True)

# Որակական սյուներ
cat_cols = df.select_dtypes(include='object').columns

# Լրացնում ենք բաց թողնված որակական արժեքները ամենահաճախ հանդիպողով
for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Check for missing values in the entire dataframe
print(df.isnull().sum())
```

# 5. Feature Engineering

```python
# Օրինակ 1․ նոր հատկանիշ՝ թվային սյուների գումար
if len(num_cols) >= 2:
    df["num_sum_feature"] = df[num_cols].sum(axis=1)

# Օրինակ 2․ թվային սյուների միջին
if len(num_cols) >= 1:
    df["num_mean_feature"] = df[num_cols].mean(axis=1)
```
